# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import normalize

%matplotlib inline
```

```python
# load data

data = np.loadtxt("3Ddata.txt")
data.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(data[:,0], data[:,1], data[:,2],c = data[:,3])
```

```python
PCAdata = data[:,range(0,3)]
PCAlabel = data[:,3]
n = len(PCAdata)
y = np.zeros((n,2)) #new data
nc = 2

NormalData = normalize(PCAdata, axis=0, norm='max')

sigma = np.cov(NormalData.T)
[w, V] = np.linalg.eigh(sigma)

# Sigma is symmetric
# Eigenvalue and Eigenvectors, small to large
# Column is an eigenvector

EigenVector = V[:,range(-1,-nc-1,-1)] # nc= The number of cluster
EigenValue = w[range(-1,-nc-1,-1)]

for i in range(0,n):
    y[i,0] = np.inner(NormalData[i,:],EigenVector[:,0]) #1st PC
    y[i,1] = np.inner(NormalData[i,:],EigenVector[:,1]) #2nd PC
```

```python
# PCA result

plt.scatter(y[:,0], y[:,1], c = PCAlabel)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Dimensionality Reduction by PCA')
```

The above result is similar with the original data plot.

Maybe, the primary PC's are related to the axis, x and z.

Therefore, the result looks reasonable.
